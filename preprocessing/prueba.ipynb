```python
import os
from collections import Counter
import matplotlib.pyplot as plt

LABELS_DIR = "../data/processed_dl/labels"
class_counter = Counter()

for label_file in os.listdir(LABELS_DIR):
    if label_file.endswith(".txt"):
        with open(os.path.join(LABELS_DIR, label_file), 'r') as f:
            for line in f:
                if line.strip():
                    class_name = line.split()[0]
                    class_counter[class_name] += 1

class_names = sorted(class_counter.keys())

counts = [class_counter[class_name] for class_name in class_names]

# Gráfica
plt.figure(figsize=(10,6))
plt.bar(class_names, counts, color='skyblue')
plt.xlabel("Clases")
plt.ylabel("Número de ejemplos")
plt.title("Distribución de clases en el dataset (detectadas automáticamente)")
plt.grid(axis='y')
plt.show()

# Mostrar las clases encontradas
print(f"Clases detectadas: {class_names}")
```
